```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn 
```

```python
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('График квадравтичной функции')
x = np.linspace(-10, 10, 1000)
y = [i ** 2 for i in x]
plt.plot(x, y, label='Квадратичная ф-я')
plt.legend()
plt.show()
```
